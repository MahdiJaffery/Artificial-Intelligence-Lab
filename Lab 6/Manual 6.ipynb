```python
#   Genetic Algorithm for N-Queens
import random

Genotype = [[random.randint(1, 8) for i in range(8)] for i in range(100)]

# Genotype = [2, 4, 7, 4, 8, 5, 5, 2]

# print(Genotype)

def getFitness(Genotype):
    maxPairs = 8 * 7 // 2
    conflict = 0
    for i in range(8):
        for j in range(i+1, 8):
            if Genotype[i] == Genotype[j]:
                conflict += 1

            if abs(Genotype[i] - Genotype[j]) == abs(i-j):
                conflict += 1
    
    return maxPairs - conflict

bestPair = list(list())

def getBestParents(Geotype):
    sortedPopulation = sorted(Genotype, key=getFitness, reverse=True)

    bestParentOne = sortedPopulation[0]
    bestParentTwo = sortedPopulation[1]

    return [bestParentOne, getFitness(bestParentOne), bestParentTwo, getFitness(bestParentTwo)]

bestParents = getBestParents(Genotype)

def getMutation(Genotype):
    bestParents = getBestParents(Genotype)
    
    crossoverPoint = random.randint(1, 7)
    ChildOneTwo = bestParents[0][:crossoverPoint] + bestParents[2][crossoverPoint:]
    ChildTwoOne = bestParents[2][:crossoverPoint] + bestParents[0][crossoverPoint:]


    # print(f"Best Parent One: {bestParents[0]}\nFitness: {bestParents[1]}")
    # print(f"Best Parent Two: {bestParents[2]}\nFitness: {bestParents[3]}")

    # print(f"Child One: {ChildOneTwo}\nFitness: {getFitness(ChildOneTwo)}")
    # print(f"Child Two: {ChildTwoOne}\nFitness: {getFitness(ChildTwoOne)}")

    return ChildOneTwo, ChildTwoOne

def solveNQueens(Genotype):
    noImprovementCount = 0
    bestFitness = 0
    
    for i in range(1000):
        Children = getMutation(Genotype)

        MutationIndexSize = int(len(Children[0]) * 0.2)
        MutationIndices = random.sample(range(8), MutationIndexSize)

        for Child in Children:
            for index in MutationIndices:
                Child[index] = random.randint(1, 8)

        Genotype[-2:] = Children  
        Genotype = sorted(Genotype, key=getFitness, reverse=True)

        if getFitness(Genotype[0]) == 28:
            print("Solution Found!")
            return Genotype[0],

        currentFitness = getFitness(Genotype[0])
        if currentFitness > bestFitness:
            bestFitness = currentFitness
            noImprovementCount = 0
        else:
            noImprovementCount += 1

        if noImprovementCount > 100:
            for i in range(2):
                Genotype[random.randint(0, 99)] = [random.randint(1, 8) for _ in range(8)]
            noImprovementCount = 0

    return Genotype[0], Genotype[1]


Children = solveNQueens(Genotype)

if len(Children) > 1:
    print("Solution Not Found!")
    print("Closest:")
    print(f"Child One: {Children[0]}\nFitness: {getFitness(Children[0])}")
    print(f"Child Two: {Children[1]}\nFitness: {getFitness(Children[1])}")
else:
    print(f"Solution: {Children[0]}\nFitness: {getFitness(Children[0])}")
```
