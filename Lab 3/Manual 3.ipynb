```python
#   Task 1
import numpy as np
from collections import deque

def BFS(graph, start, end):
    if graph[start[0]][start[1]] != 0:
        print("Start is an Obstacle!")
        return False
    if graph[end[0]][end[1]] != 0:
        print("End is an Obstacle!")
        return False
    
    visited = set()
    queue = deque([(start, [start])])

    rows, cols = np.shape(graph)

    while queue:
        current, path = queue.popleft()

        if current == end:
            return path
        if current in visited:
            continue

        visited.add(current)

        row, col = current

        for i, j in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
            newRow, newCol = row + i, col + j

            if 0 <= newRow < rows and 0 <= newCol < cols and graph[newRow][newCol] == 0:
                queue.append([(newRow, newCol), path + [(newRow, newCol)]])
            
        
    return False

def DFS(graph, start, end):
    if graph[start[0]][start[1]] != 0:
        print("Start is an Obstacle!")
        return False
    if graph[end[0]][end[1]] != 0:
        print("End is an Obstacle!")
        return False
    
    visited = set()
    
    rows, cols = np.shape(graph)

    stack = deque([(start, [start])])

    while stack:
        current, path = stack.pop()

        if current == end:
            return path
        if current in visited:
            continue

        visited.add(current)

        for i, j in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
            newRow, newCol = current[0] + i, current[1] + j

            if 0 <= newRow < rows and 0 <= newCol < cols and graph[newRow][newCol] == 0:
                stack.append([(newRow, newCol), path + [(newRow, newCol)]])
    
    return False


myList = np.ones((10, 10)).astype(int)
myList = (myList + .01 * np.random.randn(10, 10)).astype(int)

print("Graph: \n", myList)

print("\n\nBFS\n")
BFS_path = BFS(myList, (0, 0), (9, 9))

if BFS_path:
    print("Path Found!\nPath: ", BFS_path)
else:
    print("Path Not Found!")


print("\n\nDFS\n")
DFS_path = DFS(myList, (0, 0), (9, 9))

if DFS_path:
    print("Path Found!\nPath: ", DFS_path)
else:
    print("Path Not Found!")
```

```python
#   Task 2
import sys
import random
import numpy as np

def getVals(matrix, i, j):
    return matrix[i][j], matrix[i-1][j], matrix[i+1][j], matrix[i][j-1], matrix[i][j+1]

def getMinima(matrix):
    rows, cols = matrix.shape

    localMinima = []

    for i in range(1, rows - 1):
        for j in range(1, cols - 1):
            curr, up, down, left, right = getVals(matrix, i, j)

            if curr < up and curr < down and curr < left and curr < right:
                localMinima.append([i, j])

    return localMinima

rows = cols = random.randint(1, 10)
myMatrix = [[random.randint(0, 50) for i in range(cols)] for j in range(rows)]

myMatrix = np.array(myMatrix)

for row in myMatrix:
    print(row)

if rows < 3:
    print("No Local Minima")
else:
    localMinima = getMinima(myMatrix)

    if localMinima == []:
        print("No local minima found")
    else:
        for sets in localMinima:
            print(f"Local Minima at ({sets}): {myMatrix[sets[0]][sets[1]]}")
```
